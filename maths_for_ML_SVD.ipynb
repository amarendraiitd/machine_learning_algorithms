**SVD_Singular_Value_decomposition**_using_Numpy[not_scikit-learn]

```python
import numpy as np
np.set_printoptions(precision=4, suppress=True)
```

```python
#let's take 3x4 matrix
a= np.array([[1,2,3,4],[1,1,2,3],[0,1,1,0]])
a
```

```python
#number of singular values(sigma) is atmost 3,since original matrix has rank 3.
# the left and right singular matrix are orthonormal
u,s,vh= np.linalg.svd(a, full_matrices=True)
```

```python
print(u) #3x3 matrix
u.shape
```

```python
print(u.dot(u.T)) #if dot product with its transpose is identity then orthogonal.
```

```python
print(s)  #3x1 (contains singular vlaues of sigma)
s.shape
```

```python
print(vh) #4x4 matrix
vh.shape  #it is also a orthogonal matrix
```

```python
sd = np.diag(s) #makes 3x3 diagonal matrix
sd
```

```python
b= np.zeros((3,4))
b
```

```python
# replace first 3 column of b by columns of sd
b[:,:-1]=sd   #slicing method on matrix
sigma = b
b
```

```python
print(np.dot(np.dot(u,sigma),vh)) #product of u, s and vh
# this is same as matrix "a"
```

```python
## now taking only two rank approximation of "s"
s1= [6.7509, 1.1734, 0]
```

```python
sd=np.diag(s1)
sd
```

```python
b= np.zeros((3,4))
b
```

```python
b[:,:-1] = sd
sigma = b
sigma
```

```python
print(np.dot(np.dot(u,sigma),vh))
print(a)
# almost approximate to matrix "a"
```

**Image_compresion_using_low_rank_approximation** (Let's work on an Image of IIT Delhi)

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #for image display
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```python
url ="/content/download_04_nov.jpg"
myImg = io.imread(url) #3d image
gray_image = cv.cvtColor(myImg, cv.COLOR_BGR2GRAY) #converted to 2d image
img_mat = np.array(list(gray_image), float)
print(img_mat)
plt.imshow(img_mat)
img_mat.shape
```

```python
# scale the image matrix before SVD
img_mat_scaled = (img_mat-img_mat.mean())/img_mat.std()

# perform SVD using np.linalg.svd
U, s, V = np.linalg.svd(img_mat_scaled)

# compute variance explained by each singular vector of 's'
var_explained = np.round(s**2/np.sum(s**2), decimals=3)
```

```python
import seaborn as sns
# let's print 1st 10 singular vecotrs
sns.barplot(x=list(range(1,11)), y=var_explained[0:10], color='dodgerblue')
plt.xlabel('Singular Vector', fontsize=16)
plt.ylabel('Variance Explained', fontsize=16)
plt.tight_layout()
# plt.savefig('svd_scree_plot.png', dpi=150, figsize=(8,8))
plt.savefig("Line_Plot_with_Pandas_Python.jpg")
```

```python
# reconstruct with top 5 singular values
num_components = 5
reconst_img_5 = np.array(U[:, :num_components]) .dot(np.diag(s[:num_components])
 .dot(np.array(V[:num_components, :])))
plt.imshow(reconst_img_5)
```

```python
# reconstruct with top 10 singular values
num_components = 10
reconst_img_10 = np.array(U[:, :num_components]) .dot(np.diag(s[:num_components])
 .dot(np.array(V[:num_components, :])))
plt.imshow(reconst_img_10)
```

```python
# reconstruct with top 50 singular values
num_components = 50 #original image has almost 700+ singular values
reconst_img_50 = np.array(U[:, :num_components]) .dot(np.diag(s[:num_components])
 .dot(np.array(V[:num_components, :])))
plt.imshow(reconst_img_50)
```

```python
# reconstruct with top 100 singular values
num_components = 100 #original image has almost 700+ singular values
reconst_img_100 = np.array(U[:, :num_components]) .dot(np.diag(s[:num_components])
 .dot(np.array(V[:num_components, :])))
plt.imshow(reconst_img_100)
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
axs[0,0].imshow(reconst_img_5)
axs[0,0].set_title('Reconstructed Image: 5 SVs', size=16)
axs[0,1].imshow(reconst_img_10)
axs[0,1].set_title('Reconstructed Image: 10 SVs', size=16)
axs[1,0].imshow(reconst_img_50)
axs[1,0].set_title('Reconstructed Image: 50 SVs', size=16)
axs[1,1].imshow(reconst_img_100)
axs[1,1].set_title('Reconstructed Image: 100 SVs', size=16)
plt.tight_layout()
```
